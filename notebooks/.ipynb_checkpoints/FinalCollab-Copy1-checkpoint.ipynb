### <span style="color:black">__Marine Oxygen Processes__

#### __Key Terms__:
* Dissolved Oxygen (DO) = This is the amount of dissolved oxygen gas in the water. The dissolved oxygen is important for the survival of marine life. It is measured by (mg/L) or as a percent of saturation.
* Apparent Oxygen Utilization (AOU) = the difference between the saturation oxygen concentration (O2 sat) and the observed oxygen concentration in water. This helps estimate the amount of oxygen in the water through the respiration of organic matter as it cycles from the top to the bottom of the ocean
* Oxygen Saturation (O2 Saturation) = This is measured in percent. It shows the ratio of the actual dissolved oxygen to the maximum oxygen that seawater can hold at a given temperature and salinity.

### <span style="color:cornflowerblue">#___What impacts the Oceans Oxyge?_?:
* Water Temperature --> Hotter water holds less oxygen because oxygen is a gas. Cold water can hold more oxygen. For example, when you heat a Coke, it is more likely to explode because the gas is more complex to contain than a can of Coke that is cooling down.
* Wind patterns --> The existence of wind increases turbulence and enhances air mixing, which can raise oxygen levels. 
* Nutrient levels --> High-nutrient environments can lead to increased algae growth, which is known to be responsible for decreasing oxygen levels in the water. 
* Plant growth and variance --> Plants in the ocean use photosynthesis, which adds oxygen to the water when it is sunny. Therefore, some areas have more oxygen due to photosynthesis and sunlight timing. 
* Organic and inorganic matter  --> Decomposition of organic matter involves oxygen intake through biological processes, while inorganic matter can cause chemical or physical reactions. 

#### <span style="colorcronflowerblued"> __How does ocean deoxygenation impact ecosystems and humans__? 
Oxygen is essential in the ocean because marine plants and animals need it to survive, just like on land. Ocean deoxygenation is when the oxygen levels in ocean and coastal waters drop. This happens when oxygen is used up faster than it can be replaced by photosynthesis or mixing from wind and waves. Human activities like burning fossil fuels and farming are the leading causes. Warmer water holds less oxygen, making it harder for oxygen to mix into deeper ocean parts. Farming nutrient pollution can also cause algae to grow and then die off, using up oxygen in the process. These changes lead to habitat loss, fewer marine species, and the growth of low-oxygen areas called dead zones, which hurt marine life and fisheries (UC Scripps Institution of Oceanography).
